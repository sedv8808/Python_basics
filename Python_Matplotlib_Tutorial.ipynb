```python
# Load libraries
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

## Matplotlib for Python

To show how we use matplotlib, we will use the Titanic data set. This data set is known for its simplicity and is widely used for introduction purposes. 

The most important function is plot; this functions allows us to "draw" 2D data. 

```python
# Load the data
titanic = pd.read_csv("data/titanic.csv")
```

```python
# Draw a plot
titanic.plot(x = "age", y = "fare")
plt.show()
```

We can see that the above graph is not very informative. However, if we could only segment the types of fare, or make a scatter plot instead of a line, that would be easier.
For this purpose, same as geom in R's ggplot, we can change the type of graph we want to plot. Let's repeat the graph using a scatterplot instead.

```python
titanic.plot.scatter(x = "age", y = "fare")
plt.show()
```

```python
# we can give format to the plot with a few commands

class1 = titanic[titanic.pclass == 1]
class2 = titanic[titanic.pclass == 2]
class3 = titanic[titanic.pclass == 3]

plt.scatter(class1[['age']], class1[['fare']], alpha = 0.4)
plt.scatter(class2[['age']], class2[['fare']], alpha = 0.5)
plt.scatter(class3[['age']], class3[['fare']], alpha = 0.2)
plt.title('Titanic Data Set')
plt.ylabel('Fare')
plt.xlabel('Age')

plt.legend(['First Class', 'Second Class', 'Third Class'])
plt.show()
```

We could also try facetting the graphs!

```python
# we can make subplots as well using two methods. 
fig = plt.figure()
fig.suptitle("Titanic Data Set", fontsize=13)
fig.text(0.5, 0.04, 'Age', ha='center')
fig.text(0.04, 0.5, 'Fare', va='center', rotation='vertical')
    
plt.subplot(1, 3, 1)
plt.scatter(class1[['age']], class1[['fare']], alpha = 0.4)
plt.title('First Class')

plt.subplot(1, 3, 2)
plt.scatter(class2[['age']], class2[['fare']], alpha = 0.5)
plt.title('Second Class')

plt.subplot(1, 3, 3)
plt.scatter(class3[['age']], class3[['fare']], alpha = 0.2)
plt.title('Third Class')

plt.show()
```

```python
# This method will also allow us to format and change the axis and axes in a more intuitive manner.

fig, axs = plt.subplots(3, 1, figsize=(8, 8), sharex = True)

axs[0].scatter(class1[['age']], class1[['fare']])
axs[0].set_title('Class 1')
axs[1].scatter(class2['age'], class2['fare'])
axs[1].set_title('Class 2')
axs[2].scatter(class3['age'], class3['fare'])
axs[2].set_title('Class 3')


fig.text(0.5, 0.04, 'Age', ha='center')
fig.text(0.04, 0.5, 'Fare', va='center', rotation='vertical')

plt.show()
```

Matplotlib offers the user different geoms such as bar graphs, chart graphs, violin graphs among others. It also allows to plot images where we would want to conduct some work, maybe an MRI image or just another image to crop/change size. 

However, the above graphs, are just to exemplify some of the uses Matplotlib has.
